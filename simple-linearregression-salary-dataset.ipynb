# LINEAR REGRESSION MODEL TO PREDICT THE SALARY BASED ON EXPERIENCE

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('/kaggle/input/salary-dataset-simple-linear-regression/Salary_dataset.csv',index_col=0)
```

```python
df.head()
```

```python
df.describe()
```

# PLOTTING A GRAPH BETWEEN EXPERIENCE AND SALARY TO CHECK IF OUR DATA IS LINEAR

```python
plt.scatter(df['YearsExperience'],df['Salary'],color='purple')
plt.xlabel('YEARS OF EXPERIENCE')
plt.ylabel('SALARY [$]')
```

# PREPROCESSING THE DATA

```python
x = df.iloc[:,0:1]
y = df.iloc[:,1]
```

```python
x.head()
```

```python
y.head()
```

# TRAINING OUR LINEAR REGRESSION MODEL

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=3)
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.predict(x_test.iloc[0].values.reshape(1,1))
```

# PLOTTING THE GRAPH OF OUR TRAINED REGRESSION MODEL

```python
plt.scatter(df['YearsExperience'],df['Salary'],color='gold')
plt.plot(x_train,lr.predict(x_train))
plt.xlabel('YEARS OF EXPERIENCE')
plt.ylabel('SALARY [$]')
```

# CALCULATING THE SLOPE AND INTERCEPT OF THE LINE

```python
# y = mx + b [equation of a line]
m = lr.coef_ # 9418.56697369
b = lr.intercept_ # 25256.91033001527
# y = 9418.56697369 * x + 25256.91033001527
```

# CALCULATING THE ACCURACY OF THE MODEL

```python
r2_score = lr.score(x_test,y_test)
print(r2_score*100)
```
